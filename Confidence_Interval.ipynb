```python
beml_df=pd.read_csv('BEML.csv')
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
glaxo_df = pd.read_csv("GLAXO.csv")
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1)
```

```python
print('The mean of the Close Price of Glaxo Stock is:',glaxo_df.Close.mean())
print('The std deviation of the Close Price of Glaxo Stock is:',glaxo_df.Close.std())
glaxo_df_ci = stats.norm.interval(0.95,
                                  loc = glaxo_df['Close'].mean(),
                                  scale = (glaxo_df['Close'].std()/np.sqrt(len(glaxo_df['Close']))))
print( 'Close Price at 95% confidence interval is:', np.round(glaxo_df_ci, 4))
#95% of the times, we can expect the Closing Price of Glaxo stock to be within the range of [2507.2177, 2558.0458]
```

```python
print('The mean of the Glaxo Stock is:',glaxo_df.gain.mean())
print('The std deviation of the Glaxo Stock is:',glaxo_df.gain.std())
glaxo_df_ci=stats.norm.interval(0.95, loc = glaxo_df['gain'].mean(),
                                scale = (glaxo_df['gain'].std()/np.sqrt(len(glaxo_df['gain']))))
print( 'Gain at 95% confidence interval is:', np.round(glaxo_df_ci, 4))
#95% of the times, we can expect the gain from Glaxo stock to be within the range of [-0.0002,  0.001]
```

```python
print('The mean of the BEML Stock is:',beml_df.gain.mean())
print('The std deviation of the BEML Stock is:',beml_df.gain.std())
beml_df_ci = stats.norm.interval(0.99,loc=beml_df['gain'].mean(),
                                 scale=(beml_df['gain'].std()/np.sqrt(len(beml_df['gain']))))
print('Gain at 99% confidence interval is:', np.round(beml_df_ci, 4))
#99% of the times, we can expect the gain from BEML stock to be within the range of [-0.0678,0.0684]
```
